```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-20, 21, 2)
```

```python
def graph_Plotting(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'ro-', markerfacecolor ='b')
    
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

**1. Sigmoid Function**

```python
y = 1 / (1 + np.exp(-x))
graph_Plotting(y, "Sigmoid(X)")
```

**2. Tanh Function**

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
graph_Plotting(y, "TanH(X)")
```

**3. Step Function**

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
graph_Plotting(y,"ReLU(X)")
```

**4. Rectified Linear Unit (RELU) Function**

```python
y = list(map(lambda a: a if a>=0 else 0, x))
graph_Plotting(y,"ReLU(X)")
```

**5. Leaky ReLU Function**

```python
y = list(map(lambda a: a if a>=0 else a*0.1, x))
graph_Plotting(y,"Leaky ReLU(X)")
```

**6. Exponential Linear Unit (ELU) Function**

```python
alpha = 0.5
def elu(x):
    if x > 0:
        return x
    else:
        return alpha * (np.exp(x) - 1)
y = list(map(elu, x))

graph_Plotting(y,"ELU(X)")
```

**7. Scaled Exponential Linear Unit (SELU) Function**

```python
alpha = 1.6732
lambdaa = 1.05
def selu(x):
    if x >= 0:
        return lambdaa * x
    else:
        return lambdaa * alpha * (np.exp(x) - 1)
y = list(map(selu, x))
graph_Plotting(y,"SELU(X)")
```
